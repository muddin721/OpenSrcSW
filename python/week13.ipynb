```python
#실습 1 : 생일 최고/최저기온 표로 출력하기
import csv
import matplotlib.pyplot as plt

birthday = input("생일을 입력해주세요(yyyy-MM-dd) : ").split('-')
file = open('seoul_all.csv', 'r', encoding='utf-8')
data = csv.reader(file, delimiter = ',')
next(data)
temp = [(row[0].split('-'), float(row[3]), float(row[4])) for row in data if len(row) == 5 and row[3] != "" and row[4] != ""]
minTemp = [row[1] for row in temp if row[0][1] == birthday[1] and row[0][2] == birthday[2]]
maxTemp = [row[2] for row in temp if row[0][1] == birthday[1] and row[0][2] == birthday[2]]
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
plt.title("생일 {0}월 {1}일의 최저/최고기온 그래프".format(birthday[1], birthday[2]))
plt.plot(minTemp, color='b', label="최저기온")
plt.plot(maxTemp, color='r', label="최고기온")
plt.legend()
plt.show()
file.close()
```

```python
#실습 2 : 1월과 8월 최고기온 표로 출력하기
import csv
import matplotlib.pyplot as plt

file = open('seoul_all.csv', 'r', encoding='utf-8')
data = csv.reader(file, delimiter = ',')
next(data)
maxTemp = [(row[0].split('-'), float(row[4])) for row in data if len(row) == 5 and row[4] != ""]
jan = [row[1] for row in maxTemp if int(row[0][1]) == 1]
aug = [row[1] for row in maxTemp if int(row[0][1]) == 8]
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
plt.title("서울의 1월과 8월 최고기온 그래프")
plt.hist(jan, bins = 100, color='b', label="1월")
plt.hist(aug, bins = 100, color='r', label="8월")
plt.legend()
plt.show()
file.close()
```
