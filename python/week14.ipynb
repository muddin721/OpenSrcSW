```python
#문제:인구구조
import re
import csv
import math
import matplotlib.pyplot as plt
file = open('age.csv', 'r', encoding='utf-8')
data = csv.reader(file, delimiter = ',')
next(data)
pop = dict([(row[0].split()[1], [int(col.replace(",", "")) for col in row[3:]]) for row in data if len(row[0].split()) == 3])

target = "광진구"
diff = dict([[key, math.sqrt(sum([(pop[key][i] - pop[target][i]) ** 2 for i in range(101)]) / 101)] for key in pop.keys() if key != target])
maxD = max(diff.keys(), key = lambda k : diff[k])
minD = min(diff.keys(), key = lambda k : diff[k])
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
plt.style.use("ggplot")
plt.plot(pop[target], label = target)
plt.plot(pop[maxD], label = "{0}({1:.2f})".format(minD, diff[maxD]))
plt.plot(pop[minD], label = "{0}({1:.2f})".format(minD, diff[minD]))
plt.xlabel('나이')
plt.ylabel('인구 수')
plt.legend()
plt.show()
```
